# Adding lettering to subplots

example implementations of adding lettering to subplots:
    
- https://stackoverflow.com/questions/25543978/matplotlib-annotate-subplots-in-a-figure-with-a-b-c

```python
import matplotlib.pyplot as plt 
import numpy as np
import string 

from sciplotlib import style as spstyle
num_points = 100
x = np.random.normal(size=num_points)
y = np.random.normal(size=num_points)


with plt.style.context(spstyle.get_style('nature-reviews')):
    fig, axs = plt.subplots(2, 2,  sharex=True, sharey=True)
    for ax in axs.flatten():
        ax.scatter(x, y)
        
def add_lettering(axs, size=10, xloc=-0.1, yloc=1.1, weight='bold'):
    
    for n, ax in enumerate(axs.flatten()):
        
        ax.text(xloc, yloc, string.ascii_uppercase[n], 
                transform=ax.transAxes, 
                size=size, weight=weight)
    
    return axs


axs = add_lettering(axs)
    
```
