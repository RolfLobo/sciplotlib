```python
%load_ext autoreload
import matplotlib.pyplot as plt 
import numpy as np
import string 

from sciplotlib import style as spstyle
# from sciplotlib import misc as spmisc
```

## The Economist

```python
num_points = 100
x = np.random.normal(loc=3, size=num_points)
y = np.random.normal(loc=3, size=num_points)
```

```python
def add_economist_rectangle(fig, ax, xloc=0.1, yloc=1.04, width=0.05, height=0.04):
    """
    
    
    References 
    ---------
    economist color scheme: http://pattern-library.economist.com/color.html
    """
    
    economist_red = '#e3120b'
    
    fig.patches.extend([plt.Rectangle((xloc, yloc), width=width, height=height,
                              fill=True, color='#e3120b', alpha=1, zorder=1000,
                              transform=fig.transFigure, figure=fig)])
    
    return fig, ax
    
    
```

```python
def add_datasource(fig, ax, xloc=0.1, yloc=-0.1, fontsize=10, weight='light', alpha=0.8):
    
    fig.text(xloc, yloc, s='Source: IMF', fontsize=fontsize, alpha=alpha,
            weight=weight)
    
    return fig, ax
```

### Single plot 

```python
%autoreload
with plt.style.context('/home/timsit/sciplotlib/sciplotlib/stylesheets/economist.mplstyle'):
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    # ax.set_title('Figure title', loc='left')
    ax.text(0, 1.2, 'Main title', weight='bold', size=13, transform=ax.transAxes)
    ax.text(0, 1.1, 'This is the usual long subtitle', transform=ax.transAxes)
    fig, ax = add_economist_rectangle(fig, ax, xloc=0.125, yloc=1.1, width=0.05, height=0.02)
    fig, ax = add_datasource(fig, ax, xloc=0.125, yloc=0, alpha=0.6)
    plt.savefig('/home/timsit/sciplotlib/figures/economist-scatter.png', dpi=300,  bbox_inches='tight')
```

### Multiple plots 


```python
%autoreload
import matplotlib as mpl 
with plt.style.context('/home/timsit/sciplotlib/sciplotlib/stylesheets/economist.mplstyle'):
    fig, axs = plt.subplots(1, 3)

    
    
    fig.set_size_inches(9, 3)
    for n, ax in enumerate(axs):
        ax.scatter(x, y)
        ax.set_title('Subtitle %s' % str(n), loc='left')

    fig.text(0.1, -0.1, s='Source: IMF', fontsize=10, alpha=0.8)
    
    fig.patches.extend([plt.Rectangle((0.1,1.04), width=0.05, height=0.04,
                                  fill=True, color='#e3120b', alpha=1, zorder=1000,
                                  transform=fig.transFigure, figure=fig)])
    
```
